```python
kmers = [dna[i:i + k] for i in range(len(dna)-k + 1)]
kmerfreq = Counter(kmers)
```

```python
max_kmers = kmerfreq.most_common()
```

```python
max_value = max_kmers[0][1]
```

```python
for i in max_kmers:
    if i[1] >= max_value:
        print(i[0])
    else:
        break
```

```python
kmer_counts = defaultdict(int)
max_kmer = []
max_kmer_count = 0
for i in range(len(dna) - k + 1):
    kmer = dna[i:i + k]
    kmer_counts[kmer] += 1
    
    if kmer_counts[kmer] > max_kmer_count:
        max_kmer = [kmer]
        max_kmer_count = kmer_counts[kmer]
        
    elif kmer_counts[kmer] == max_kmer_count:
        max_kmer.append(kmer)

print("\n".join(max_kmer))
```
